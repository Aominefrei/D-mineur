# **Training avec fit() pour l'Humidité**

```
m = Prophet()
```

```
m.fit(data_humidity_2017)
```

# **Prediction pour l'Humidité**

### *make_future_dataframe()*

```
future = m.make_future_dataframe(periods=31)
future.tail()
```

### *Predict()*

```
forecast = m.predict(future)
```

```
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

# **Plot pour l'Humidité**

```
fig1 = m.plot(forecast)
```

```
fig2 = m.plot_components(forecast)
```

# **Interprétation de l'Humidité**

*   Les prédictions obtenues sur le dataset représentant l'humidité sont on ne peut plus cohérentes. On peut remarquer que l'humidité varie selon les saisons; en effet, en été, l'humidité baisse considérablement par rapport aux autre saison. Inversement, on remarque des pics sur notre graphique sur les mois d'hiver. On peut donc facilement comprendre et interpréter les différentes saisons d'activitées des abeilles.

# **Training avec fit() pour la Température**

```
m = Prophet()
m.fit(data_temperature_2017)
```

# **Prediction pour la Température**

### *make_future_dataframe()*

```
future = m.make_future_dataframe(periods=31)
future.tail()
```

### *Predict()*

```
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

# **Plot pour la Température**

```
fig3 = m.plot(forecast)
fig4 = m.plot_components(forecast)
```

# **Interprétation de la Température**

Les prédictions obtenues sur le dataset représentant la température sont on ne peut plus cohérentes. On peut remarquer que la température varie selon les saisons; en effet, en été, la température augmente considérablement par rapport aux autre saison. Inversement, on remarque des descentes de température sur notre graphique sur les mois d'hiver. On peut donc facilement comprendre et interpréter les différentes saisons d'activitées des abeilles.
